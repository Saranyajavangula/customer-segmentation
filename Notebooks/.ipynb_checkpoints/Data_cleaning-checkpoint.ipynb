```python
#cleaning the data 
query="""select ot.invoice,
                  ot.stock_code,
             CASE
                when description IS NULL then 'Unknown'
                else description
                end as description,
                ot.price * ot.quantity  as total_cost,
                ot.invoice_date,
                ot.price,
                ot.quantity,
                ot.customer_id,
                ot.country
              from online_transactions ot
              left join (select * 
                          from stock_description 
                            where description <> '?' ) sd on ot.stock_code = sd.stock_code
             where  ot.customer_id <> ''
             and ot.stock_code not in ('BANK CHARGES', 'POST', 'D', 'M', 'CRUK')"""
```

```python
ot_clean=pd.read_sql(query,conn)
```

```python
ot_clean
```

```python
#finding duplicates 
duplicates = ot_clean[ot_clean.duplicated(keep=False)]
```

```python
duplicates
```

```python
#removing the duplicates 
ot_with_out_duplicates = ot_clean.drop_duplicates(keep='first')
```

```python
ot_with_out_duplicates
```

```python
ot_with_out_duplicates.to_sql(name='online_transactions_cleaned',con=conn, index=False, if_exists="replace")
```

```python
# definition of missing data 

def identify_missing_data(df):
    """
    This function is used to identify missing data
    
    @param df pandas DataFrame
    
    @return a DataFrame with the percentage of missing data for every feature and the data types
    """
    
    percent_missing = df.isnull().mean()
    
    missing_value_df = pd.DataFrame(percent_missing).reset_index() # convert to DataFrame
    missing_value_df = missing_value_df.rename(columns = {"index" : "variable",
                                                                0 : "percent_missing"}) # rename columns

    missing_value_df = missing_value_df.sort_values(by = ['percent_missing'], ascending = False) # sort the values
    
    data_types_df = pd.DataFrame(df.dtypes).reset_index().rename(columns = {"index" : "variable",
                                                                0 : "data_type"}) # rename columns
    
    missing_value_df = missing_value_df.merge(data_types_df, on = "variable") # join the dataframe with datatype
    
    missing_value_df.percent_missing = round(missing_value_df.percent_missing*100, 2) # format the percent_missing
    
    return missing_value_df[missing_value_df.percent_missing > 0]
```

```python
missing_data = identify_missing_data(ot_with_out_duplicates)
```
