```python
#How many invoices does the table contain 
query=""" select count(invoice) as total_invoices 
            from online_transactions_cleaned"""
```

```python
# when was the first and last purchase 
query= """ select min(invoice_date) as first_purchase,
                  max(invoice_date) as last_purchase 
                  from online_transactions_cleaned"""
```

```python
# how many customers does the table contain 
query= """ select count(distinct customer_id) as number_of_customers 
                  from online_transactions_cleaned"""
```

```python
pd.read_sql(query,conn)
```

```python
# How many different types of stocks did the customer purchase?
query= """ select count(distinct stock_code) as types_of_stocks
                  from online_transactions_cleaned"""
```

```python
pd.read_sql(query,conn)
```

### Market Analysis

```python
# which country has high sales (Top 10)
query = """select country,
                  sum(quantity) as product_sold
           from online_transactions_cleaned 
           group by country 
           order by product_sold desc
           limit 10 """
```

```python
top10_countries=pd.read_sql(query,conn)
```

```python
top10_countries
```

```python
# What is the most popular stock in UK? (top 10 sold items - does this differ across markets?)
query = """ select stock_code,
                   description,
                   country,
                   sum(quantity) as product_sold
                   from online_transactions_cleaned
                   where country='United Kingdom'
                   group by stock_code
                   order by product_sold desc
                   limit 10"""         
```

```python
top10_uk=pd.read_sql(query,conn)
```

```python
top10_uk
```

```python
# What is the most popular stock in Netherlands?
query = """ select stock_code,
                   description,
                   country,
                   sum(quantity) as product_sold
                   from online_transactions_cleaned
                   where country='Netherlands'
                   group by stock_code
                   order by product_sold desc
                   limit 10"""  
```

```python
top10_nl=pd.read_sql(query,conn)
```
