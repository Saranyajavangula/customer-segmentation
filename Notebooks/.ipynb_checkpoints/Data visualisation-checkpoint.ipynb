```python
# convering invoice date field into datetime
ot_transactions['invoice_date']= pd.to_datetime(ot_transactions['invoice_date'])
```

```python
ot_transactions.dtypes
```

```python
# extracting date,month,year,day from datetime field
ot_transactions['Date'] = ot_transactions['invoice_date'].dt.date
ot_transactions['Year'] = ot_transactions['invoice_date'].dt.year
ot_transactions['Month'] = ot_transactions['invoice_date'].dt.month
ot_transactions['Day'] = ot_transactions['invoice_date'].dt.day
```

```python
ot_transactions
```

```python
# no of invoices per month 
invoices_per_year_month = ot_transactions.groupby(['Year', 'Month']).size().to_frame(name='InvoiceCount')
```

```python
invoices_per_year_month.reset_index(inplace=True)
```

```python
invoices_per_year_month
```

```python
plt.figure(figsize=(10, 6))
for year in invoices_per_year_month['Year'].unique():
    year_data = invoices_per_year_month[invoices_per_year_month['Year'] == year]
    plt.plot(year_data['Month'], year_data['InvoiceCount'], marker='o', label=f'Year {year}')

plt.title('Number of Invoices per Year and Month')
plt.xlabel('Month')
plt.ylabel('Number of Invoices')
plt.xticks(range(1, 13))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

###  Calculate Frequency
Determine the purchase frequency for each customer. This can be done by counting the number of transactions for each customer over a specific period.

### Define Segmentation Criteria
Decide on the criteria for segmenting your customers based on frequency. Common segments include:

High-frequency customers: Customers who purchase very often.
Medium-frequency customers: Customers who purchase regularly but not as often as high-frequency customers.
Low-frequency customers: Customers who purchase infrequently.
Inactive customers: Customers who have not made a purchase in a long time.

```python
### Implement Segmentation 
# Calculate purchase frequency
frequency = ot_transactions.groupby('customer_id').size().reset_index(name = 'Frequency')
frequency
```

```python
# Define segmentation bins
bins = [0, 50, 100, float('inf')]
labels = ['Low-frequency', 'Medium-frequency', 'High-frequency']
# Segment customers
frequency['Segment'] = pd.cut(frequency['Frequency'], bins=bins, labels=labels, right=False)
```

```python
frequency
```

```python
customer_seg_frequency=frequency.groupby('Segment')['customer_id'].count().to_frame(name='value').reset_index()
```

```python
customer_seg_frequency
```

```python
customer_seg_frequency.plot.bar(x='Segment', y='value',rot=45)
```
